```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Data/Combined.csv")
```

```python
df.head()
```

```python
df[["Home Score", "Away Score"]] = df["Score"].str.split('-', 1, expand = True)
```

```python
df.head()
```

```python
def check_result(row):
    if row["Home Score"] > row["Away Score"]:
        return 1
    elif row["Home Score"] < row["Away Score"]:
        return -1
    else:
        return 0
```

```python
df["Result"] = df.apply(check_result, 1)
```

```python
df.to_csv("Data/Match Data.csv", index = False)
```
